```python
def contador_vogais(exagerado):

    vogais = {
        'a': 0, 'à': 0, 'á': 0, 'â': 0, 'ã': 0,
        'e': 0, 'é':0, 'ê':0,
        'i': 0, 'í': 0,
        'o': 0, 'ó': 0, 'ô': 0, 'õ': 0,
        'u': 0, 'ú': 0
        }
    
    for char in exagerado.lower():
       if char in vogais:
         vogais[char] += 1
    return vogais

def ler_arq(arquivo):
  with open(arquivo, 'r', encoding='utf-8') as arquivo:
    return arquivo.read()
  
arquivo = 'exagerado_letra.txt'

conteudo = ler_arq(arquivo)

result = contador_vogais(conteudo)

print("Contagem de vogais no arquivo: ")
for vogal, contagem in result.items():
   print(f"{vogal}: {result}")
```
